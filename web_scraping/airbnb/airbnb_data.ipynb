This Python code is designed to scrape data from Airbnb listings in Markham, Ontario, Canada. It performs web scraping by sending requests to Airbnb's website, extracting information such as listing names, prices per night in Canadian dollars, and descriptions, and then organizing this data into a structured format. The code first collects URLs for specific listings from search result pages and then visits each listing to retrieve the desired details. Ultimately, it compiles the scraped data into a Pandas DataFrame, making it accessible and ready for analysis or further processing. This code essentially automates the process of gathering Airbnb listing information for Markham, facilitating tasks like price analysis or comparison

```python
url='https://www.airbnb.ca/s/Markham--Ontario--Canada/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&monthly_start_date=2023-10-01&monthly_length=3&price_filter_input_type=0&price_filter_num_nights=5&channel=EXPLORE&query=Markham%2C%20ON&place_id=ChIJqUwyoO_V1IkRYz2yLIFSPfc&date_picker_type=calendar&source=structured_search_input_header&search_type=autocomplete_click'
r = requests.get(url)
soup = BeautifulSoup(r.text,'lxml')


for i in range(1, 4):
    np = soup.find('a', class_ = 'l1ovpqvx c1ytbx3a dir dir-ltr').get('href')
    cnp = 'https://www.airbnb.ca/' + np
    #print(cnp)
    
    url = cnp
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'lxml')
    
Name=[]
Price=[]
Description=[]


url='https://www.airbnb.ca/s/Markham--Ontario--Canada/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&monthly_start_date=2023-10-01&monthly_length=3&price_filter_input_type=0&price_filter_num_nights=5&channel=EXPLORE&query=Markham%2C%20ON&place_id=ChIJqUwyoO_V1IkRYz2yLIFSPfc&date_picker_type=calendar&source=structured_search_input_header&search_type=autocomplete_click'
r = requests.get(url)
soup = BeautifulSoup(r.text,'lxml')


for i in range(1, 3):
    np = soup.find('a', class_ = 'l1ovpqvx c1ytbx3a dir dir-ltr').get('href')
    cnp = 'https://www.airbnb.ca/' + np
    #print(cnp)

    url = cnp
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'lxml')

    names = soup.find_all('div', class_ = 't1jojoys dir dir-ltr')
    for i in names:
        i =i.text
        Name.append(i)
        

    prices = soup.find_all('div', class_ = '_1jo4hgw')
    for i in prices:
        n = i.text.strip()
        cn = re.search(r'\$\d+', n)
        if cn:
            price_str = cn.group(0)  # Get the price string (e.g., '$100')
            price_f = float(price_str[1:])  # Convert to float, removing the '$' symbol
            Price.append(price_f)

    description = soup.find_all('span', class_ = 't6mzqp7 dir dir-ltr')
    for i in description:
        i= i.text
        Description.append(i)
   
    
        
df= pd. DataFrame({'Name': Name, 'Price/night(CAD)': Price, 'Description':Description})
print('data downloaded')
```
